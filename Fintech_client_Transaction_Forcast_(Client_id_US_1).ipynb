```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt

# Définissez la fonction create_sequences
def create_sequences(data, sequence_length):
    sequences = []
    for i in range(len(data) - sequence_length):
        seq = data[i:i+sequence_length]
        sequences.append(seq)
    return np.array(sequences)

# Charger les données
df = pd.read_excel('/content/dailly.xlsx')

# Triez le DataFrame par date
sorted_df = df.sort_values('Date')

# Définissez la colonne 'Date' comme indice
df.set_index('Date', inplace=True)

# Normalisation des données
scaler = MinMaxScaler()
df['amount_transaction'] = scaler.fit_transform(df['amount_transaction'].values.reshape(-1, 1))

# Créez des séquences pour l'entraînement du modèle
sequence_length = 10
train_data = create_sequences(df['amount_transaction'].values, sequence_length)

X_train = train_data[:, :-1]
y_train = train_data[:, -1]

# Reshape des données
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))

# Divisez les données en ensembles d'entraînement et de validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Construisez le modèle LSTM avec des ajustements
model1 = Sequential()
model1.add(LSTM(units=200, input_shape=(X_train.shape[1], 1), return_sequences=True))
model1.add(Dropout(0.2))  # Ajout de dropout pour régularisation
model1.add(LSTM(units=50))
model1.add(Dropout(0.2))
model1.add(Dense(units=1, activation='linear'))  # Vous pouvez essayer d'autres fonctions d'activation

# Compilez le modèle avec un optimiseur différent
model1.compile(optimizer='rmsprop', loss='mean_squared_error')

# Entraînez le modèle avec un nombre d'époques plus élevé
history = model1.fit(X_train, y_train, epochs=200, batch_size=32, validation_data=(X_val, y_val))

# Faites des prédictions sur l'ensemble d'entraînement
train_pred = model1.predict(X_train)

# Créez un index de dates pour les prédictions
pred_dates = pd.date_range(start=df.index[sequence_length], periods=len(train_pred), freq='D')

# Inversez la normalisation pour les prédictions
train_pred = scaler.inverse_transform(train_pred)
y_train = scaler.inverse_transform(y_train.reshape(-1, 1))

# Visualisez les prédictions sur les données d'entraînement
plt.plot(pred_dates, y_train, label='Valeurs réelles')
plt.plot(pred_dates, train_pred, label='Prédictions')
plt.xlabel('Date')
plt.ylabel('Amount Transaction')
plt.legend()
plt.show()

```

```python
# Utiliser le modèle LSTM pour générer des prédictions en continuant à nourrir le modèle avec ses propres prédictions
last_sequence = X_train[-1]  # Utilisez le dernier point de vos données d'entraînement
predicted_values = []

# Faire des prédictions sur la période de deux mois en utilisant les prédictions précédentes comme entrée
for _ in range(40):  # Pour deux mois
    prediction = model1.predict(np.array([last_sequence]))  # Prédire avec le dernier point

    # Ajouter un peu d'aléatoire aux prédictions pour éviter la convergence vers une valeur
    prediction_with_noise = prediction + np.random.normal(0, 0.1)  # Ajouter un peu de bruit

    # Ajouter la prédiction à la liste uniquement si elle est positive
    if prediction_with_noise[0, 0] > 0:
        predicted_values.append(prediction_with_noise[0, 0])

    last_sequence = np.roll(last_sequence, -1)  # Décaler la séquence d'une position
    last_sequence[-1] = prediction_with_noise  # Utiliser la prédiction avec bruit comme entrée

# Dénormaliser les prédictions si vous avez appliqué une normalisation
predicted_values = scaler.inverse_transform(np.array(predicted_values).reshape(-1, 1))

# Générer les valeurs temporelles pour la période de deux mois à partir du 7 novembre 2022
start_date = pd.to_datetime('2022-11-07')
future_dates = pd.date_range(start=start_date, periods=len(predicted_values), freq='D')

# Tracer les prédictions pour la période de deux mois
plt.plot(future_dates, predicted_values, label='Prédictions')

plt.xlabel('Temps')
plt.ylabel('Valeur des transactions')
plt.title('Prédictions des transactions pour deux mois')
plt.legend()
plt.show()
```

```python

# Create a DataFrame for the predictions and the corresponding dates
predicted_df = pd.DataFrame({'Date': future_dates, 'Prediction': predicted_values.flatten()})
```
